```
stemmer = PorterStemmer()
x = x.apply(stemmer.stem)
```

```
vector = TfidfVectorizer()
x = vector.fit_transform(x)
```

```
print(x)
```

```
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity between the vectorized text data
similarity_matrix = cosine_similarity(x)

def get_top_n_similar_items(item_index, similarity_matrix, n=10):
    """
    Gets the top N most similar items for a given item index.

    Args:
        item_index (int): The index of the item for which to find similar items.
        similarity_matrix (np.ndarray): The cosine similarity matrix.
        n (int): The number of top similar items to return.

    Returns:
        list: A list of indices of the top N most similar items (excluding the item itself).
    """
    # Get the similarity scores for the item at item_index
    similarity_scores = list(enumerate(similarity_matrix[item_index]))

    # Sort the items based on similarity scores in descending order
    sorted_similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Get the indices of the top N similar items (excluding the item itself)
    top_n_indices = [i[0] for i in sorted_similarity_scores[1:n+1]]

    return top_n_indices

# Example usage (finding top 10 similar items for the first item)
# You can change the index (0 in this case) to find similar items for other items.
# top_10_similar_items = get_top_n_similar_items(0, similarity_matrix, n=10)
# print("Top 10 similar items for the first item:", top_10_similar_items)
```
